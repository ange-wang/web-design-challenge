```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df=pd.read_csv('assets/cities.csv')
```

```python
df.head()
```

```python
with open('assets/to_html.html', 'w') as w:
    w.write(df.to_html(index=False))
```

```python
ax = df.plot.scatter('Lat','Max Temp', figsize=(8,6), s=20, c='#037E8C')
ax.set(facecolor="#DEDDD9", xlabel='Latitude', ylabel="Max Temperature (F)", title= "Latitude vs. Max Temperature (10/05/17)")
plt.xlim(-70,90)
plt.ylim(0,120)
plt.rcParams.update({'font.size': 30})
plt.savefig("assets/images/fig1.svg")
```

```python
ax = df.plot.scatter('Lat','Humidity', figsize=(8,6), s=20, c='#037E8C')
ax.set(facecolor="#DEDDD9", xlabel='Latitude', ylabel="Humidity (%)", title= "Latitude vs. Humidity (10/05/17)")
plt.xlim(-70,90)
plt.ylim(0,120)
plt.savefig("assets/images/fig2.svg")
```

```python
ax = df.plot.scatter('Lat','Cloudiness', figsize=(8,6), s=20, c='#037E8C')
ax.set(facecolor="#DEDDD9", xlabel='Latitude', ylabel="Cloudiness (%)", title= "Latitude vs. Cloudiness (10/05/17)")
plt.xlim(-70,90)
plt.ylim(-10,110)
plt.savefig("assets/images/fig3.svg")
```

```python
ax = df.plot.scatter('Lat','Wind Speed', figsize=(8,6), s=20, c='#037E8C')
ax.set(facecolor="#DEDDD9", xlabel='Latitude', ylabel="Wind Speed (mph)", title= "Latitude vs. Wind Speed (10/05/17)")
plt.xlim(-70,90)
plt.ylim(-3,35)
plt.savefig("assets/images/fig.svg")
```
